```python
import pandas as pd
import json
from tinydb import TinyDB, Query
import sqlalchemy as db
```

```python
df_tmp = pd.read_feather("spiele_db.ftr")

df_tmp
```

```python
engine = db.create_engine('sqlite://', echo=False)
```

```python
df_tmp.to_sql('games', con=engine)
engine.execute("SELECT * FROM games").fetchall()
```

```python
metadata = db.MetaData()

games_table = db.Table('games', metadata, autoload=True, autoload_with=engine)

print(metadata.tables)
print(games_table.columns)
print(games_table.metadata)
```

```python
game_db = TinyDB('spiele_tinydb.json')
```

```python
for i in df_tmp.index:
    # print(df_tmp.loc[i].to_json())
    game_db.insert(json.loads(df_tmp.loc[i].to_json()))
```

```python
my_query = Query()

# game_db.search((my_query.game_rating.exists()) & (my_query.game_rating >= 90))

test_func = lambda s: True if(isinstance(s, float) and s >= 85) else False
tmp_result = game_db.search(my_query.game_rating.test(test_func))
```

```python
def search_by_rating(min_rating = 0, max_rating = 100):
    my_query = Query()
    test_func = lambda s: True if(isinstance(s, float) and s >= min_rating and s <= max_rating) else False
    return game_db.search(my_query.game_rating.test(test_func))

print(search_by_rating(95))
```

```python
print(type(tmp_result)) # a list of results
print(len(tmp_result)) # the number of results
print(type(tmp_result[0])) # the type of the first result (a tinyDB document, i.e. a dictionary and a doc_id)
print(tmp_result[0].doc_id)
print(tmp_result[0]["game_title"]) # pne column / attribute of the first result
```

```python
tmp_result_sorted = sorted(tmp_result, key=lambda x: x["game_rating"], reverse=True)
```

```python
for item in tmp_result_sorted:
    print(item["game_title"], ": ", round(item["game_rating"], 2), " / ", item["game_rating_count"])
```
